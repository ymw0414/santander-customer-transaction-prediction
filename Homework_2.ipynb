# Data Preprocessing

```python
# Change working directory to the absolute path of the shallow utilities folder
%cd $abspath_util_shallow

# Import the shallow utitilities
%run pmlm_utilities_shallow.ipynb
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# =============================================
# Loading the data
# =============================================

# Read train and test csv files from the Santander competition
df_raw_train = pd.read_csv(abspath_curr + 'data/train.csv', header=0)
df_train = df_raw_train.copy(deep=True)

df_raw_test = pd.read_csv(abspath_curr + 'data/test.csv', header=0)
df_test = df_raw_test.copy(deep=True)

# target column is binary: 0 means no transaction, 1 means transaction
target = 'target'

print("df_train shape:", df_train.shape)
print("df_test shape:", df_test.shape)
display(df_train.head())

# =============================================
# Splitting the data
# =============================================

# using 80/20 split same as case study
df_train, df_val = train_test_split(df_train, train_size=0.8, random_state=random_seed)
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)

# saving these before concat so the slicing doesn't get messed up later
n_train = len(df_train)
n_val = len(df_val)

# =============================================
# Handling uncommon variables
# =============================================

# test data doesn't have 'target' column so need to check what's common
df_common_var = common_var_checker(df_train, df_val, df_test, target)

uncommon_in_train = np.setdiff1d(df_train.columns, df_common_var['common var'])
uncommon_in_val   = np.setdiff1d(df_val.columns,   df_common_var['common var'])
uncommon_in_test  = np.setdiff1d(df_test.columns,  df_common_var['common var'])

print("Uncommon in train:", list(uncommon_in_train))

df_train = df_train.drop(columns=uncommon_in_train)
df_val   = df_val.drop(columns=uncommon_in_val)
df_test  = df_test.drop(columns=uncommon_in_test)

# =============================================
# Handling identifiers
# =============================================

df = pd.concat([df_train, df_val, df_test], sort=False)

# ID_code is just a row identifier, not useful for prediction
df_id = id_checker(df)
print("Identifiers found:", list(df_id.columns))

df_train = df_train.drop(columns=np.intersect1d(df_id.columns, df_train.columns))
df_val   = df_val.drop(columns=np.intersect1d(df_id.columns, df_val.columns))
df_test  = df_test.drop(columns=np.intersect1d(df_id.columns, df_test.columns))

# =============================================
# Handling missing data
# =============================================

df = pd.concat([df_train, df_val, df_test], sort=False)
df_nan = nan_checker(df)

# target column shows up as missing because test data doesn't have labels - exclude it
df_nan = df_nan[df_nan['var'] != target].reset_index(drop=True)
print("Missing values:")
display(df_nan)

# checked the data and Santander doesn't really have missing values,
# but keeping this here just to follow the same pipeline structure as case study
df_miss = df_nan[df_nan['dtype'] == 'float64'].reset_index(drop=True)

df_train = df.iloc[:n_train, :].reset_index(drop=True)
df_val   = df.iloc[n_train:n_train + n_val, :].reset_index(drop=True)
df_test  = df.iloc[n_train + n_val:, :].reset_index(drop=True)

if len(df_miss) > 0:
    si = SimpleImputer(missing_values=np.nan, strategy='mean')
    df_train[df_miss['var']] = si.fit_transform(df_train[df_miss['var']])
    df_val[df_miss['var']]   = si.transform(df_val[df_miss['var']])
    df_test[df_miss['var']]  = si.transform(df_test[df_miss['var']])

# =============================================
# Handling categorical variables
# =============================================

df = pd.concat([df_train, df_val, df_test], sort=False)
df_cat = cat_var_checker(df)
print("Categorical variables:")
display(df_cat)

# all 200 features (var_0 ~ var_199) are already numeric so nothing to encode here
if len(df_cat) > 0:
    cat_features = list(np.setdiff1d(df_cat['var'], [target]))
    df = pd.get_dummies(df, columns=cat_features)
    df_train = df.iloc[:n_train, :].reset_index(drop=True)
    df_val   = df.iloc[n_train:n_train + n_val, :].reset_index(drop=True)
    df_test  = df.iloc[n_train + n_val:, :].reset_index(drop=True)

# =============================================
# Splitting features and target
# =============================================

features = [col for col in df_train.columns if col != target]

X_train = df_train[features].values
X_val   = df_val[features].values
X_test  = df_test[features].values

# using LabelEncoder because the submission code uses le.inverse_transform later
# cast to int first since concat with test data (no target) turns the column into float64
le = LabelEncoder()
y_train = le.fit_transform(df_train[target].values.astype(int))
y_val   = le.transform(df_val[target].values.astype(int))

print("X_train:", X_train.shape, " X_val:", X_val.shape, " X_test:", X_test.shape)

# =============================================
# Scaling the features
# =============================================

# MinMaxScaler to put everything in [0, 1] range, which works better for neural networks
mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
X_val   = mms.transform(X_val)
X_test  = mms.transform(X_test)
```

# Hyperparameter Tuning

```python
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

# =============================================
# Creating the dictionary of the models
# =============================================

# MLPClassifier is the main model for this homework (shallow neural network)
models = {
    'mlp': MLPClassifier(early_stopping=True, random_state=random_seed)
}

# =============================================
# Creating the dictionary of the pipelines
# =============================================

pipes = {}
for name, model in models.items():
    pipes[name] = Pipeline([('model', model)])

# =============================================
# Getting the predefined split cross-validator
# =============================================

X_train_val, y_train_val, ps = get_train_val_ps(X_train, y_train, X_val, y_val)

# =============================================
# Creating the dictionary of the parameter grids
# =============================================

param_grids = {}

# for MLP I tried different hidden layer sizes to find a good architecture
# also tuning alpha (regularization) and learning rate
# (100, 50) and (200, 100) are two-layer networks, (100,) is single layer
param_grids['mlp'] = [
    {
        'model__hidden_layer_sizes': [(100,), (100, 50), (200, 100)],
        'model__alpha': [0.0001, 0.001, 0.01],
        'model__learning_rate_init': [0.001, 0.01]
    }
]

# =============================================
# Creating the directory for the cv results
# =============================================

directory = os.path.dirname(abspath_curr + 'result/cv_results/GridSearchCV/')
if not os.path.exists(directory):
    os.makedirs(directory)

# =============================================
# Tuning the hyperparameters
# =============================================

# using f1_macro as specified in the assignment instructions
best_score_params_estimator_gs = []

for name in pipes.keys():
    print(f"Tuning {name}...")

    gs = GridSearchCV(estimator=pipes[name],
                      param_grid=param_grids[name],
                      scoring='f1_macro',
                      n_jobs=2,
                      cv=ps,
                      return_train_score=True)

    gs.fit(X_train_val, y_train_val)

    best_score_params_estimator_gs.append([gs.best_score_, gs.best_params_, gs.best_estimator_])

    # save cv results to csv
    cv_results = pd.DataFrame.from_dict(gs.cv_results_).sort_values(by=['rank_test_score', 'std_test_score'])

    important_columns = ['rank_test_score',
                         'mean_test_score',
                         'std_test_score',
                         'mean_train_score',
                         'std_train_score',
                         'mean_fit_time',
                         'std_fit_time',
                         'mean_score_time',
                         'std_score_time']

    cv_results = cv_results[important_columns + sorted(list(set(cv_results.columns) - set(important_columns)))]
    cv_results.to_csv(abspath_curr + 'result/cv_results/GridSearchCV/' + name + '.csv', index=False)

# sort by best score, higher f1_macro is better
best_score_params_estimator_gs = sorted(best_score_params_estimator_gs, key=lambda x: x[0], reverse=True)

pd.DataFrame(best_score_params_estimator_gs, columns=['best_score', 'best_params', 'best_estimator'])
```

# Model Selection

```python
# Best model = highest validation f1_macro from the grid search
best_score_gs, best_params_gs, best_estimator_gs = best_score_params_estimator_gs[0]

print(f"Best validation f1_macro: {best_score_gs:.6f}")
print(f"Best parameters: {best_params_gs}")
print(f"Best estimator: {best_estimator_gs}")

print(f"\nBaseline: 0.670522")
print(f"Beat baseline: {best_score_gs > 0.670522}")
```
